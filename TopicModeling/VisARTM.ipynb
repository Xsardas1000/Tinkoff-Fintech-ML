```python
import pandas as pd
import tqdm
import artm
import numpy as np
```

```python
import pickle
```

```python
import pyLDAvis
```

```python
pyLDAvis.__version__
```

```python
VIS_MODEL = 'vtb24/vis/vis_model_replies/'
```

```python
Phi = pd.read_pickle(VIS_MODEL + 'model_replies_phi.pkl')
```

```python
Phi.shape
```

```python
vocab = Phi.index.tolist()
vocab = [word.encode('utf-8') for word in vocab]
```

```python
Theta = pd.read_pickle(VIS_MODEL + 'model_replies_theta.pkl')
```

```python
Theta.shape
```

```python
with open(VIS_MODEL + 'doc_lengths.pkl', 'rb') as f:
    doc_lengths = pickle.load(f)
```

```python
len(doc_lengths)
```

```python
with open(VIS_MODEL + 'vocab.pkl', 'rb') as f:
    words = pickle.load(f)
```

```python
term_frequency = [words[word] for word in vocab]
```

```python
len(term_frequency)
```

```python
model_vis_data = pyLDAvis.prepare(doc_topic_dists=Theta.transpose(),
                                  topic_term_dists=Phi.transpose(), 
                                  doc_lengths=doc_lengths, 
                                  term_frequency=term_frequency,
                                  vocab=vocab)
```

```python
pyLDAvis.display(model_vis_data)
```
